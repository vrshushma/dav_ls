1.To create a Dataframe.

```python
import pandas as pd
data = {"calories":[420,380,390],
        "duration":[50,40,45]}
df = pd.DataFrame(data)
print(df)
```

2.To locate the rows in a DataFrame

```python
print(df.loc[[0,1]])
```

3.Named indexes

```python
df = pd.DataFrame(data,index=["day1","day2","day3"])
print(df)
```

4.To locate named index

```python
print(df.loc["day2"])
```

```python
print(df.loc["day3"])
```
