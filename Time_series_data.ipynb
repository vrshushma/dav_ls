trend

```python
import pandas as pd
import matplotlib.pyplot as plt

data = {
     "Date": pd.to_datetime(["2022-01-01", "2022-02-01", "2022-03-01", "2022-04-01", "2022-05-01"]),
     "Value":[10, 15, 22, 28, 35]
 }

df = pd.DataFrame(data)

df.set_index("Date", inplace=True)


plt.figure(figsize=(10, 6))
plt.plot(df.index,  df["Value"], marker="o")
plt.xlabel("Date")
plt.ylabel("Value")
plt.title("Time Series with Trend")
plt.grid(True)
plt.show()
```

seasonality

```python
import pandas as pd
import matplotlib.pyplot as plt

data = {
     "Date":pd.to_datetime (["2022-01-01", "2922-02-01", "2022-03-01", "2022-04-01", "2822-05-01"]),
     "Sales": [100, 110, 120, 135, 150]
}

df = pd.DataFrame(data)
plt.figure(figsize=(10, 6))
plt.plot(df["Date"], df("Sales"), label="Daily Data")
plt.xlabel("Date")
plt.ylabel("Sales")
plt.title("Time Series with Seasonality")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
df = pd.read_csv('https://raw.githubusercontent.com/selva86/datasets/master/a10.csv', parse_dates=['date'], index_col='date')

def plot_df(df, x, y, title="", xlabel='Date', ylabel='Value', dpi=100):
 plt.figure(figsize=(16,5))
 plt.plot(x, y, color='red')
 plt.gca().set(title=title, xlabel=xlabel, ylabel=ylabel)
 plt.show()

plot_df(df, x=df.index, y=df.value, title='Monthly anti-diabetic drug sales in Australia from 1992 to 2008.')
```
