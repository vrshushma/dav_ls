3.textdata visualization

```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud
from collections import Counter

text_data = "Python is a popular programming language used for data analysis and visualization. It is easy to learn and has a large community of users Python libraries such as Matplotlib, Seaborn, and Plot"

wordcloud = WordCloud(width=800, height=400, background_color='black').generate(text_data)

plt.imshow(wordcloud)
plt.figure(figsize=(10, 5))
plt.title('Word Cloud')
plt.show()

words = text_data.split()
word_counts = Counter(words)

plt.figure(figsize=(10, 5))
plt.bar(word_counts.keys(), word_counts.values())
plt.title('Histogram of Word Frequencies')
plt.xlabel('Words')
plt.ylabel('Frequency')
plt.xticks (rotation=90)
plt.show()
```

4. stopwords

```python
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')

def remove_stop_words(string) :
    stop_words = set(stopwords.words('english'))
    words = string.split()
    filtered_words = [word for word in words if word.lower() not in stop_words]
    new_string = ' '.join(filtered_words)
    return new_string


input_string = "This is an example sentence to remove stop words from. "
result = remove_stop_words(input_string)
print("Original string:", input_string)
print("Modified string:", result)
```
